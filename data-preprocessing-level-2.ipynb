```python
# 2. Extract standard sector names from messy strings
# Yeh code automatically 'b block sector 43' ko 'sector 43' bana dega
def clean_sector_names(text):
    if pd.isna(text):
        return text
    
    # Check if 'sector' is in the text
    match = re.search(r'(sector\s*-?\s*\d+[a-zA-Z]?)', str(text), re.IGNORECASE)
    if match:
        # Standardize format: "sector 43", "sector 89a"
        cleaned = match.group(1).replace('-', ' ').replace('  ', ' ').lower()
        return cleaned
    return text

# Apply the regex cleaning function
df['sector'] = df['sector'].apply(clean_sector_names)

# Optional: Drop outside cities/junk data entirely
out_of_bounds = ['lucknow', 'farukhnagar', 'sultanpur', 'bhora kalan', 'gopal nagar, garhi ghasita, sonipat', ', delhi ncr', 'damdma', 'new']
df['sector'] = df['sector'].replace(out_of_bounds, pd.NA)
```
